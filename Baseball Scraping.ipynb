```python
import random
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from bs4 import BeautifulSoup
from bs4 import Comment
import requests
import re
```

```python
url = 'https://www.baseball-reference.com/leagues/majors/2021-advanced-pitching.shtml'
response = requests.get(url)
page = response.text.replace("<!--","").replace("-->","")
soup = BeautifulSoup(page, "lxml")
```

```python
player_table = soup.find_all('table')[-1]
```

```python
headers = player_table.find('thead').find_all('tr')[-1]
```

```python
column_names = [col.get_text() for col in headers.find_all('th')]
column_names
```

```python
df_2021 = pd.DataFrame(columns=column_names)
df_2021
```

```python
table_body = player_table.find('tbody').find_all('tr')
```

```python
stats = []
for i in range(len(table_body)-1):
    stats.append(table_body[i].find_all('th')[0].text)
    for j in range(len(table_body[i].find_all('td'))):
        stats.append(table_body[i].find_all('td')[j].text)
```

```python
stats = list(filter(('Rk').__ne__, stats))
```

```python
idx=0
for a in range(0,len(stats),21):
    df_2021.loc[idx] = [stats[a],stats[a+1],stats[a+2],stats[a+3],
                        stats[a+4],stats[a+5],stats[a+6],stats[a+7],
                        stats[a+8],stats[a+9],stats[a+10],stats[a+11],
                        stats[a+12],stats[a+13],stats[a+14],stats[a+15],
                        stats[a+16],stats[a+17],stats[a+18],stats[a+19],
                        stats[a+20]]
    idx +=1

df_2021.tail(50)
```
